```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from imblearn.datasets import make_imbalance
from sklearn.datasets import make_blobs, make_classification
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.metrics import classification_report
from prettytable import PrettyTable
```

```python
#1) Synthetic data generation

def synthetic_gen(dimension, size=10000, classes=2, imbalance=0.97):
    X, y = make_classification(
    n_samples=size,
    n_features=dimension,
    n_informative=1,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[imbalance],
    flip_y=0,
    random_state=42
)
    return X, y
```

```python
num_classes = 2
size = 1000
dim = 2
imbalance = 0.90

X, y = synthetic_gen(dim, size, num_classes, imbalance)
plt.scatter(X[:, 0][y == 0], X[:, 1][y == 0])
plt.scatter(X[:, 0][y == 1], X[:, 1][y == 1])
plt.tight_layout()
print(X.shape, y.shape)
```

```python
#2) Stratified split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#3) Random Forest Model (basic)
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(max_depth=2, random_state=0)
rf_clf.fit(X_train, y_train)
```

```python
#3.1) [Basic Random Forest] : Train Confusion Matrix

y_train_hat = rf_clf.predict(X_train)

cm_train = confusion_matrix(y_train, y_train_hat)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=rf_clf.classes_, yticklabels=rf_clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('[Random Forest (basic)] : Confusion Matrix of train')
plt.show()
```

```python
#3.2) [Basic Random Forest] : Train Accuracy, Recall, Precision, F1


basic_rf_train_report = classification_report(y_train, y_train_hat, output_dict=True)
accuracy = accuracy_score(y_train, y_train_hat)
table = PrettyTable()
table.field_names = ["Class", "Precision", "Recall", "F1-Score", "Support"]

for class_name, metrics in basic_rf_train_report.items():
    if class_name.isdigit(): 
        table.add_row([class_name, metrics['precision'], metrics['recall'], metrics['f1-score'], metrics['support']])
table.add_row(["Accuracy", "", "", accuracy, ""])

print(table)
```

```python
#3.3) [Basic Random Forest] : Test Confusion Matrix

y_test_hat = rf_clf.predict(X_test)

cm_test = confusion_matrix(y_test, y_test_hat)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=rf_clf.classes_, yticklabels=rf_clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('[Random Forest (basic)] : Confusion Matrix of test')
plt.show()
```

```python
#3.4) [Basic Random Forest] : Test Accuracy, Recall, Precision, F1

basic_rf_report = classification_report(y_test, y_test_hat, output_dict=True)
accuracy = accuracy_score(y_test, y_test_hat)
table = PrettyTable()
table.field_names = ["Class", "Precision", "Recall", "F1-Score", "Support"]

for class_name, metrics in basic_rf_report.items():
    if class_name.isdigit(): 
        table.add_row([class_name, metrics['precision'], metrics['recall'], metrics['f1-score'], metrics['support']])
table.add_row(["Accuracy", "", "", accuracy, ""])

print(table)
```

```python
import pandas as pd

def create_table_row(id, classifier_name, modifiers, sampling, dataset_size, imbalance_percentage,
                     training_precision, training_accuracy, training_recall, training_f1, training_balanced_accuracy, training_size,
                     test_precision, test_accuracy, test_recall, test_f1, test_balanced_accuracy, test_size):
    
    data = {
        'ID': [id],
        'Classifier name': [classifier_name],
        'Modifiers': [modifiers],
        'Sampling': [sampling],
        'Dataset_size': [dataset_size],
        '% imbalance': [imbalance_percentage],
        'Training': {
            'Precision': [training_precision],
            'Accuracy': [training_accuracy],
            'Recall': [training_recall],
            'F1': [training_f1],
            'Balanced_accuracy': [training_balanced_accuracy],
            'Training_size': [training_size]
        },
        'Test': {
            'Precision': [test_precision],
            'Accuracy': [test_accuracy],
            'Recall': [test_recall],
            'F1': [test_f1],
            'Balanced_accuracy': [test_balanced_accuracy],
            'Test_size': [test_size]
        }
    }

    return pd.DataFrame.from_dict(data, orient='columns')


row1 = create_table_row(1, 'RandomForest', 'N/A', 'N/A', 1000, 10, 0.85, 0.75, 0.80, 0.82, 0.77, 800, 0.80, 0.72, 0.75, 0.78, 0.74, 200)
row2 = create_table_row(2, 'BalancedRandomForest', 'None', 'BalancedRandomForest', 1500, 20, 0.90, 0.80, 0.85, 0.87, 0.82, 1200, 0.88, 0.78, 0.82, 0.85, 0.80, 300)
row3 = create_table_row(3, 'SMOTE-RandomForest', 'SMOTE', 'SMOTE', 2000, 15, 0.88, 0.78, 0.82, 0.85, 0.80, 1600, 0.86, 0.76, 0.80, 0.83, 0.78, 400)

table = pd.concat([row1, row2, row3], ignore_index=True)

print(table)

# RandomForest NA NA train
# +----------+--------------------+--------------------+--------------------+---------+
# |  Class   |     Precision      |       Recall       |      F1-Score      | Support |
# +----------+--------------------+--------------------+--------------------+---------+
# |    0     | 0.975254730713246  | 0.9896602658788775 | 0.9824046920821115 |  677.0  |
# |    1     | 0.8888888888888888 | 0.7671232876712328 | 0.8235294117647058 |   73.0  |
# | Accuracy |                    |                    |       0.968        |         |
# +----------+--------------------+--------------------+--------------------+---------+


# Random Forest NA NA test
# +----------+--------------------+--------------------+--------------------+---------+
# |  Class   |     Precision      |       Recall       |      F1-Score      | Support |
# +----------+--------------------+--------------------+--------------------+---------+
# |    0     | 0.9781659388646288 |        1.0         | 0.988962472406181  |  224.0  |
# |    1     |        1.0         | 0.8076923076923077 | 0.8936170212765957 |   26.0  |
# | Accuracy |                    |                    |        0.98        |         |
# +----------+--------------------+--------------------+--------------------+---------+



```
